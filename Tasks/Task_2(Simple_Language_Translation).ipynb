```python
from googletrans import Translator, LANGUAGES

#Create a function for text translation
def translate_text(text, target_language='en'):
    translator = Translator()
    translation = translator.translate(text, dest=target_language)
    return translation.text

#Printing supported language
def print_supported_languages():
    print("Supported languages:")
    for lang_code, lang_name in LANGUAGES.items():
        print(f"{lang_code}: {lang_name}")

if __name__ == "__main__":
    print("Welcome to Simple Language Translator!")
    print_supported_languages()

    while True:
        text_to_translate = input("\nEnter text to translate (or 'q' to quit): ")

        if text_to_translate.lower() == 'q':
            print("Exiting...")
            break

        target_lang = input("Enter target language code : ")

        try:
            translated_text = translate_text(text_to_translate, target_language=target_lang)
            print(f"\nTranslated text: {translated_text}")
        except Exception as e:
            print(f"Translation failed. Error: {str(e)}")
```
